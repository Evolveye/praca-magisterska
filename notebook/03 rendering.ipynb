# World rendering

## Methods of buffer management

1. **Direct Mapping** -- Direct mapping from host memory to gpu memory. Simplest, basic method.
2. **Staging buffer** -- Mapping from host to staging buffer, then copying staging into target buffer. Faster for reading from by GPU
3. **Ring Buffer** -- Big buffer with data for several frames. It reduces the need for buffers updating
4. **Double/Tripple/N Buffering** -- storing several buffers and commiting it to GPU one by one. When first is rendering, second may be generated by CPU

Double Buffering seems to be the best option, for rapidly updating terrain
